```python
from sklearn.datasets import make_classification
from sklearn.datasets import make_moons
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve, auc

from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn import svm

import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def make_meshgrid(X, h=.01):
    """
    Create a mesh grid based on vectors X and Y.
    Date: 31 May, 2023
    Source: https://github.com/fenna/BFVM23DATASCNC5/blob/main/Exercises/E_LR_SVM.ipynb
    """
    if X.shape[1] > 2:
        raise ValueError('Requires a matrix of 2 features for visualization purposes.')
    
    x_min = X.min() - 1
    x_max = X.max() + 1
        
    x = np.arange(x_min, x_max, h)
        
    xx, yy = np.meshgrid(x, x)
    
    return xx, yy
```

```python
class SVC:
    '''
    Implements Support Vector Classification (SVC) using the 
    scikit-learn library and trains and visualises the performance.
    
    Attributes:
        X: the input features
        y: the target labels
        length: length of the input features provided
        X_train, X_test, y_train, y_test: subsets of testing and 
            training data
        
    Methods:
        train_model: trains and visualises the SVC model using an 
        optional kernel parameter ('linear', 'poly', 'rbf', 'sigmoid'). 
        If not provided the default is 'rbf'.
    '''
    
    def __init__(self, X, y):
        
        self.X = X
        self.y = y
        self.length = int(len(X) * 0.5)
        self.X_train = X[:self.length]
        self.X_test = X[self.length:]
        self.y_train = y[:self.length]
        self.y_test = y[self.length:]
        
    def train_model(self, kernel= 'rbf'):
        
        # Setting up and fitting the model
        svc = svm.SVC(kernel = kernel)
        svc.fit(self.X_train, self.y_train)
        
        # Create meshgrid
        xx, yy = make_meshgrid(self.X)
        grid = np.c_[xx.ravel(), yy.ravel()]
        
        # Getting predictions
        svc_prediction = svc.predict(grid).reshape(xx.shape)
        
        # Get support vectors
        support_vectors = svc.support_vectors_
        indices = svc.support_

        ## Visualise
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5), layout="constrained")
        fig.suptitle(f'SVC kernel: {kernel}')
        
        # Visualising SVM boundary
        axes[1].contourf(xx, yy, svc_prediction, cmap=plt.cm.Spectral)
        axes[1].scatter(X_test[:, 0], X_test[:, 1], c=y_test, edgecolor='white', linewidth=1, s=50, cmap=plt.cm.Spectral)
        axes[1].set(aspect='equal',
                   xlim=(X.min() - 1, X.max() + 1),
                   ylim=(X.min() - 1, X.max() + 1),
                   xlabel='$X_1$',
                   ylabel='$X_2$')
        axes[1].set_title('SVC boundary')

        # Visualising support vectors
        axes[0].scatter(support_vectors[:, 0], support_vectors[:, 1], c=y[indices], edgecolor='white', linewidth=1, s=50, cmap=plt.cm.Spectral)
        axes[0].set(aspect='equal',
                   xlim=(X.min() - 1, X.max() + 1),
                   ylim=(X.min() - 1, X.max() + 1),
                   xlabel='$X_1$',
                   ylabel='$X_2$')
        axes[0].set_title('SVC support vectors')
        
        plt.show()
```

```python
# Import moons dataset
X, y = make_moons(n_samples=500, noise=0.05, shuffle=True)

# SVC model
svc_model = SVC(X,y)
kernel_list = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernel_list:
    svc_model.train_model(kernel)
```

### SVM kernels
SVM kernels transform data into the desired form. The different kernels use different function to achieve this transformation:

1. **linear kernel** 

    $ K(x_{i},x_{j}) = (x_{i} * x_{j}) $
    

2. **polynomial kernel** 

    $ K(x_{i},x_{j}) = (x_{i} * x_{j} + 1)^{d} $
    
    
3. **radial basis function (rbf) kernel** 

    $ K(x_{i},x_{j}) = exp(- \sigma   || x_{i} - x_{j} ||^{2}) $
   
   
4. **sigmoid kernel** 

    $ K(x,y) = tanh(ax^{T}y + c) $

```python
# Loading data
import yaml

def get_config():
    with open("config.yaml", 'r') as stream:
        config = yaml.safe_load(stream)
    return config

config = get_config()
data = (config['breast-cancer'])
df = pd.read_csv(data)
```

```python
# Displaying number of malign
print(f'\nNumber of malignent tumors: {len(df[df.diagnosis == "M"])}')
print(f'Number of benign tumors: {len(df[df.diagnosis == "B"])}\n')
    
# Displaying number of columns and rows, datatypes, missing values
print(df.info())
```

```python
# Checking for any outliers
def check_outliers_zcore(df, threshold=3):
    
    outlier_dict = {}
    
    print('Outliers detected:')
    
    for x in df.columns.tolist():
        z_scores = stats.zscore(df[x])
        outliers = np.where(np.abs(z_scores) > 3)[0]
        
        if len(outliers) > 0:
            outlier_dict[x] = outliers.tolist()
            print(f'{x}\n---------------------- number of outliers: {len(outliers)}')
            
    return outlier_dict
```

```python
# Checking for outliers in the malignent catagory
malignent = df[df['diagnosis'] == "M"]
features = malignent.drop(['id', 'diagnosis'], axis = 1)
outliers_malign = check_outliers_zcore(features)
```

```python
# Checking for outliers in the benign catagory
benign = df[df['diagnosis'] == "B"]
features = benign.drop(['id', 'diagnosis'], axis = 1)
outliers_benign = check_outliers_zcore(features)
```

The features have varying ranges and should be scaled. 

```python
X = df.drop(['id', 'diagnosis'], axis = 1)

# Scaling the features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

The diagnosis (y) should be converted from catergorical variables to a numerical representation.

```python
y = df['diagnosis']

# Encoding the "diagnosis" column
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

### 5. Modeling (cross-validation and training)

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42)
```

```python
# Creating multiple models with different hyperparameters
models = []
# Linear regression models
lr_params = [{'penalty': 'l1', 'solver': 'liblinear'}, 
             {'penalty': 'l2', 'solver': 'liblinear'},
             {'penalty': 'l2', 'solver': 'lbfgs'},
             {'penalty': 'l2', 'solver': 'newton-cg'}]

for params in lr_params:
    lr_model = LogisticRegression(**params)
    lr_model.fit(X_train, y_train)
    models.append(lr_model)
    
# SVM
# SVM models with different hyperparameters
svm_params = [{'kernel': 'linear', 'C': 1.0},
              {'kernel': 'rbf', 'C': 1.0},
              {'kernel': 'rbf', 'C': 10.0}]

for params in svm_params:
    svm_model = SVC(**params, probability=True)
    svm_model.fit(X_train, y_train)
    models.append(svm_model)
```

### 6. Evaluation

```python
def model_eval(model):
    
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]
    
    print(f"Model: {model.__class__.__name__}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    
    # ROC-AUC curve
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    print(f"ROC-AUC: {roc_auc:.4f}")
    
    # Precision-Recall curve
    precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
    pr_auc = auc(recall, precision)
    print(f"PR-AUC: {pr_auc:.4f}")
    
    print("--------------------\n")
```

```python
# combinging parameter dict
params = lr_params + svm_params

for i in range(len(models)):
    print(f'Model {i + 1}')
    print(f'Hyperparameters used: {params[i]}')
    print("--------------------\n")
    model_eval(models[i])
```
