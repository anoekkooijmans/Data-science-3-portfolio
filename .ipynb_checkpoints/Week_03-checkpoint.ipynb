## Anomaly Detection

Anomaly detection in machine learning refers to the process of identifying patterns or instances in a dataset that deviate significantly from the norm or expected behavior. Anomalies, also known as outliers, are data points that are rare, unusual, or don't conform to the typical patterns observed in the majority of the data. Anomaly detection algorithms aim to automatically detect these anomalies. Anomaly detection in machine learning has various applications across domains, including predictive maintenance, and healthcare monitoring. By automatically identifying anomalies, these algorithms help in detecting potential risks.

```python
import yaml
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import GridSearchCV
```

```python
def get_config():
    with open("config.yaml", 'r') as stream:
        config = yaml.safe_load(stream)
    return config

def load_data(x):
    config = get_config()
    data = (config[x])
    df = pd.read_csv(data)
    return df
```

```python
df = load_data("pump-sensor")
```

```python
df.shape
```

```python
df.describe().transpose()
```

```python
#missing values
percentage_missing = df.isnull().sum().sort_values(ascending=False)/len(df)*100
print(percentage_missing)
# drop low quality columns
df.drop(['sensor_15', 'sensor_50'],inplace = True,axis=1)
# set timestamp to date
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index('timestamp')
```

```python
# Plot
df.plot(subplots =True, sharex = True, figsize = (20,50));
```

```python
# Extract machine statust
broken_rows = df[df['machine_status']=='BROKEN']
recovery_rows = df[df['machine_status']=='RECOVERING']
normal_rows = df[df['machine_status']=='NORMAL']

# use mean of the column to handle missing values and remove label in feature matrix X
m, n = df.shape
X = df.iloc[:,:n-1] # ignore machine status columns
X = X.fillna(X.mean())
X.shape
```

```python
# Scaling
scaler=StandardScaler()
X = scaler.fit_transform(X)
```

## Try to find the best configuration for one of the tested algorithms: Local Outlier Factor (LOF)

LOF is able to identify outliers by taking the local density of a point and comparing it with the local density of its k-neighbours. If the local density of the point is significantly lower than its k-neigbours the point will be considered an outlier. Each outlier gets assigned a score indicating the how much of an outlier it is. 

```python
outliers_fraction = 1 - (len(normal_rows)/(len(df))) 
algorithm = LocalOutlierFactor(contamination=outliers_fraction, n_jobs = -1)
y_pred = algorithm.fit_predict(X)

df["Local Outlier Factor"] = y_pred
print('-'*100)
print(f'number of anomolies detected')
print(df[f'{"Local Outlier Factor"}'].value_counts())
print('-'*100)
```

To optimise the configurations of LOF the number of k-neighbours needs to be set. To optimise a range from 5-15 will be tested. The contamination is set to the proportion of outliers in the dataset. This will be tested on a range of 0.05-0.15 and the default setting 'auto' where the threhsold is calculated in the function itself 

```python
warnings.filterwarnings("ignore")
# Optimising paramters
param_grid = {
    'n_neighbors': [5, 10, 15],
    'contamination': [0.05, 0.1, 0.15, 'auto']
}
grid_search = GridSearchCV(LocalOutlierFactor(n_jobs=-1), param_grid, scoring='accuracy',cv=5)
grid_search.fit(X, df['machine_status'])
# Extact optimal parameters
best_params = grid_search.best_params_
print("Best Configuration:", best_params)
```

```python
# Apply optimised model to data
algorithm = LocalOutlierFactor(
    contamination=best_params['contamination'],
    n_neighbors=best_params['n_neighbors'],
    n_jobs=-1
)
y_pred = algorithm.fit_predict(X)
```
